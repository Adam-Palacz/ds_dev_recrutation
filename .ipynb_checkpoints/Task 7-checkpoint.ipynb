#### 7. Pandas DataFrame
Write a program that will create Pandas DataFrame object based on the contensts of the table below with an index containing names of provinces:

![table](task_7_table.png)

then: \
a. Sort the data by population in descending order. \
b. Move the “province capital” column to the last place. \
c. Modify names of provinces so that they begin with a capital letter. \
d. Create a series object containing names of provinces in the index and a boolean value that indicates whether the average population per km² is greater than 140. \
e. Delete the row containing „lubuskie” province. \
f. Use pandas.DataFrame.describe() function and show statistical parameters for the analyzed data set. 

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'province': ['dolnośląskie', 'kujawsko-pomorskie', 
                    'lubelskie', 'lubuskie',
                    'łódzkie', 'małopolskie',
                    'mazowieckie', 'wielkopolskie'],
    'register nr': [2, 4, 6, 8, 10, 12, 14, 30],
    'province capital': ['Wrocław', 'Bydgoszcz / Toruń', 'Lublin',
                           'Gorzów Wielkopolski / Zielona Góra\u00b2', 'Łódź',
                          'Kraków', 'Warszawa', 'Poznań'],
    'area [km\u00b2]': [19946.70, 17971.34, 25122.46, 13987.93, 18218.95, 15182.79, 35558.48, 29826.50],
    'population': [2904198, 2086210, 2139726, 1018084, 2493603, 3372618, 5349114, 3475323]
})
```

```python
df
```

a. Sort the data by population in descending order.

```python
df.sort_values(['population'], ascending = False)
```

b. Move the “province capital” column to the last place.

```python
df = df[['province', 'register nr', 'area [km\u00b2]', 'population', 'province capital']]
df
```

c. Modify names of provinces so that they begin with a capital letter.

```python
df['province'] = df['province'].str.title()
df
```

d. Create a series object containing names of provinces in the index and a boolean value that indicates whether the average population per km² is greater than 140.

```python
import numpy as np
```

```python
df['mean_population'] = df['population']/df['area [km\u00b2]']
df
```

```python
population_sr = pd.Series(np.where(df['mean_population'] > 140, True, False), index=df['province'])
df = df.drop(columns='mean_population')
population_sr
```

e. Delete the row containing „lubuskie” province.

```python
df = df.loc[df.province != 'Lubuskie']
df
```

f. Use pandas.DataFrame.describe() function and show statistical parameters for the analyzed data set.

```python
df.describe()
```

```python
# mean value
df.mean()
```

```python
# sample standard deviation
df.std()
```
